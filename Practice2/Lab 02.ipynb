### 1.2 Create a simple dataset:
- Name    | Age | Salary
- Alice   | 25  | 50000
- Bob     | 30  | 60000
- Charlie | 35  | 75000
- David   | 22  | 45000

```python
df = pd.DataFrame(
    {
        'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 22],
        'Salary': [50000, 60000, 75000, 45000]
    }
)
```

1.3 Display the dataset:


```python
df.head()
```

1.4 Check data types of columns:


```python
df.dtypes
```

1.5 Convert 'Age' to float:

```python
df['Age'] = df['Age'].astype('float')
```

```python
df['Age'].dtypes
```

1.6 Save the dataset to CSV:


```python
df.to_csv('modified.csv', index=False)
```

## Part 2: Data Visualization
2.1 Import Matplotlib library:

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(style="whitegrid")
sns.set_palette("viridis")
plt.rcParams["axes.titleweight"] = "bold"
```

2.2 Plot a simple line chart of Name and Age 

```python
plt.plot(df['Name'], df['Age'])
plt.savefig('line_chart.png')

plt.title("Line Chart of Name & Age")
plt.xlabel("Name")
plt.ylabel("Age")

plt.show() 
```

2.3 Create a bar chart for Salary:


```python
plt.bar(df['Name'], df['Salary'])
plt.savefig('bar_chart.png')

plt.title("Bar Chart of Name & Salary")
plt.xlabel("Name")
plt.ylabel("Salary")

plt.show()
```

2.4 Scatter plot for Age and Salary:


```python
plt.scatter(df['Age'], df['Salary'])
plt.savefig('scatter_plot.png')

plt.title("Scatter Plot for Age & Salary")
plt.xlabel("Age")
plt.ylabel("Salary")

plt.show()
```
