# 2. Generate and plot data: (2)
- x = [1, 2, 3, 4, 5]
- y = [3, 5, 7, 2, 1]

### Plot the data as a scatter plot

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

plt.scatter(x, y)

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

# 3. Customize plot elements: (4)

- Change the plot market and color
- Add grid lines 
- Add proper Labels and Title

```python
plt.scatter(x, y, edgecolor='black', s=100, alpha=0.4)

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

# 4. Create different chart types: (4)
- Line Plot
- Bar Plot
- Histogram

```python
plt.plot(x, y, color='blue', marker='o')

plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
plt.bar(x, y, color='blue', alpha=0.6)

plt.title('Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
data = x+y
plt.hist(data, bins=5, color='blue', alpha=0.6)

plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# 5. Enhance plot with annotations: (4)
- Add an annotation on a random point in plot

```python
plt.plot(x, y, color='blue', marker='o')

plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

max_y = max(y)
max_index = y.index(max_y)

plt.text(x[max_index], y[max_index], f'Max Value: {max_y}', fontsize=10, color='red', ha='left')

plt.show()
```

# 6. Subplots and multiple plots: (4)
Create 2 Subplots
+ Bar Plot
+ Scatter Plot

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].bar(x, y, color='blue', alpha=0.6)
axs[0].set_title('Bar Plot')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].grid(True)

axs[1].scatter(x, y, edgecolor='black', s=100, alpha=0.4)
axs[1].set_title('Scatter Plot')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].grid(True)

plt.savefig('plot.png')
plt.tight_layout()
plt.show()
```

# 8. Color theory and application: (4)

##### Explore Color Palattes
Assign new variables
- b = range(100)
- a = [i**2 for i in x]

Create a Scatter Plot with Viridis Colormap

```python
b = range(100)
a = [i**2 for i in b]

plt.scatter(b, a, c=a, cmap='viridis')

plt.title('Scatter Plot with Viridis Colormap')
plt.xlabel('b')
plt.ylabel('a')
plt.colorbar(label='a^2')
plt.grid(True)
plt.show()
```

#### Apply Color Maps to Data
Generate a 10x10 heatmap of random data using the 'coolwarm' colormap and display the color bar with labeled data values.


```python
data = np.random.rand(10, 10)

# Interpolation helps fill in the gaps between these discrete data points.
plt.imshow(data, cmap='coolwarm', interpolation='nearest')

plt.colorbar(label='Data Values')

plt.title('Heatmap of Random Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# 9. Layout and composition: (4)

Explore the impact of different plot sizes and aspect ratios by creating a small square plot and a larger wide plot using the 'scatter' function.
- Small = 5x5
- Large = 10x5

```python
plt.figure(figsize=(5, 5))
plt.scatter(x, y, color='blue', alpha=0.6)
plt.title('Small Square Plot (5x5)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='red', alpha=0.6)
plt.title('Large Wide Plot (10x5)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

Modify the legend placement to the upper left and adjust the positioning of the annotation for the point (2, 5) in a scatter plot

```python
plt.scatter(x, y, label='Data', color='blue')

# Annotation for point (2, 5)
plt.annotate('(2, 5)', xy=(2, 5), xytext=(2.5, 5.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10)

plt.legend(loc='upper left')

plt.title('Scatter Plot with Annotation and Legend')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)
plt.show()
```

Enhance the use of whitespace in a plot by adjusting the spacing around the plot area and creating a scatter plot with improved whitespace.

```python
plt.figure(figsize=(8, 6))
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

plt.scatter(x, y, label='Data', color='blue')

plt.title('Scatter Plot with Improved Whitespace')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)
plt.show()
```

Choose appropriate chart type for the following data

# 11. Customize plot styles: (4)

Experiment with line styles, marker shapes, and color gradients:
- Dashed Line
- Triangle Marker (Green)

Utilize color gradients to represent data using a specified colormap, for example, 'YlGnBu', to create a scatter plot with a color bar.

```python
plt.plot(x, y, linestyle='--', marker='^', color='green', label='Line Plot')

colors = np.linspace(0, 1, len(x))
plt.scatter(x, y, c=colors, cmap='YlGnBu', label='Scatter Plot')
plt.colorbar(label='Color Gradient')

plt.title('Line Plot and Scatter Plot with Color Gradient')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.grid(True)
plt.show()
```
